```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
M= np.array([[1.0, 0.5, 0.0], [0.5, 1.0, -0.5],
[1.0, -0.5, 0.5],[0.0, 0.5, -0.5]])
Z = linkage (M, method='single', metric='cosine')
M_labels=["a", "b", "c", "d"]
plt.title("CAH-Exemple Simple avec color_threshold=0")
dendrogram(Z,labels =M_labels,orientation='left', color_threshold=0)
plt.show()
plt.title ("CAH- Exemple Sinple avec color_threshold=0.2")
dendrogram(Z,labels=M_labels, orientation='left', color_threshold=0.2)
plt.show()
plt.title ("CAH-Exemple Sinple avec color_threshold=@. 4")
dendrogram(Z,labels =M_labels, orientation='left', color_threshold=0.4)
plt.show()
```

```python
import pandas
fromage = pandas.read_table("C:/Users/HP/Desktop/Machine Learning/fromage.txt",sep="\t", header=0,index_col=0)
#print (fromage) i print (fromage. shape)
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
#gnrer La matrice des Liens
Z=linkage(fromage,method='ward', metric='euclidean')
dendrogram(Z,labels=fromage.index, orientation='left', color_threshold=750)
plt.title("CAH") ; plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
#matrialisation des 4 classes (hauteur t 300)
plt.title('CAH avec matrialisation des 4 classes')
dendrogram(Z,labels=fromage.index, orientation='left', color_threshold=350)
plt.show()
#dcoupage  Lo hauteur t= 300 => identifiants de 4 groupes obtenus
groupes_cah = fcluster (Z, t=350, criterion='distance')
print (groupes_cah)
#index tris des groupes
import numpy as np
idg = np.argsort(groupes_cah)
#af tcnage des o0servations et Leurs groupes
print(pandas.DataFrame(fromage.index[idg] ,groupes_cah[idg] ) )
plt.show()
```
