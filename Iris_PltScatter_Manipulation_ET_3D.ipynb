```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x =iris.data
y =iris.target
names= list(iris.target_names)

print(f'x contient {x.shape[0]} lignes et {x.shape[1]} colonnes')
print(f'il ya {np.unique(y).size} classes')
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.xlabel('longueur sepal')
plt.ylabel('largeur sepal')
plt.show()
```
