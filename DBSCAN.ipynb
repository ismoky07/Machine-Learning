```python
from sklearn.datasets import make_blobs
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas
n_sampl = 4000
random_st = 170
X,y = make_blobs(n_samples=n_sampl, random_state = random_st, cluster_std= 0.5)
X = pandas.DataFrame(X, columns = ['abscisse','ordonnée'])
from sklearn.cluster import DBSCAN
y_pred = DBSCAN(eps = 0.4, min_samples = 5, metric='manhattan').fit_predict(X)
plt.scatter(X['abscisse'],X['ordonnée'],c = y_pred);
y_pred = list(y_pred)
X=np.array(X)
for i in range(4000):
    if(y_pred[i]==-1):
        print(X[i])
```
